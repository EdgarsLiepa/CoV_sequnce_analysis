# Parse Brewery: Prediction of protein structural features result files

Before parsing the files remove description text at the start and end of the file.

Example: Input file should look like this [data/struct/alpha.txt](./allprot0429/data/struct/alpha.txt)

```python
with open('spikeprot0429/data/struct/alpha.txt') as f:
    lines = f.readlines()

data = {}
data['SEQ'] = ""
data['SS3'] = ""
data['SS3_confidence'] = ""
data['SS8'] = ""
data['SS8_confidence'] = ""
data['SA'] = ""
data['SA_confidence'] = ""
data['TA'] = ""
data['TA_confidence'] = ""
data['CD'] = ""
data['CD_confidence'] = ""
count = 0
for line in lines:
    count += 1
    if(count%13 == 1):
        data['SEQ'] += line[5:-1]
    if(count%13 == 2):
        data['SS3'] += line[5:-1]
    if(count%13 == 3):
        data['SS3_confidence'] += line[5:-1]
    if(count%13 == 4):
        data['SS8'] += line[5:-1]
    if(count%13 == 5):
        data['SS8_confidence'] += line[5:-1]
    if(count%13 == 6):
        data['SA'] += line[5:-1]
    if(count%13 == 7):
        data['SA_confidence'] += line[5:-1]
    if(count%13 == 8):
        data['TA'] += line[5:-1]
    if(count%13 == 9):
        data['TA_confidence'] += line[5:-1]
    if(count%13 == 10):
        data['CD'] += line[5:-1]
    if(count%13 == 11):
        data['CD_confidence'] += line[5:-1]

f.close()

print(data)
```
